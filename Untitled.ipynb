```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('output.csv', index_col=0)
```

```python
df['Transaction Date and Time']=pd.to_datetime(df['Transaction Date and Time'])
```

```python
df.groupby('SKU ID')\
.resample("Y", on='Transaction Date and Time')\
.count()\
.plot(kind='line')
```

```python
#time series from 2018-2020, top 5 items sold by quantity, daily transaction sold
# repeat for top 5 items by revenue
```

```python
df[(df['Transaction Date and Time']>='01-01-2018')&\
   (df['Transaction Date and Time']<='12-31-2020')]\
.resample('D', on='Transaction Date and Time').count()
```

```python
top_5 = list(df.groupby('SKU ID')['Quantity'].count().sort_values().iloc[0:5].index)
top_5
```

```python
df[(df["SKU ID"].isin(top_5)) &(df['Transaction Date and Time']>='01-01-2018')&\
   (df['Transaction Date and Time']<='12-31-2020')]\
.groupby('SKU ID')\
.resample('Y', on='Transaction Date and Time')\
.count()['Quantity']
```
